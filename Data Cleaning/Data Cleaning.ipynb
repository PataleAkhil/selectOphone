```python
#Importing libraries
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('smartphones.csv')
```

```python
df
```

From above we see that the data is pretty messy, let's clean it up.

```python
df['Product_name'] = df['Product_name'].str.split(',',expand=True)[0] +')'

df['Brand'] = df['Product_name'].str.split(' ',expand=True)[0]

df['After_Discount_Price'] = df['After_Discount_Price'].str.split('₹',expand=True)[1]

df['After_Discount_Price'] = df['After_Discount_Price'].str.replace(',','').values

df['RAM'] = df['RAM'].str.split('GB',expand=True)[0]

df['ROM'] = df['ROM'].str.split('GB',expand=True)[0]

df['Rear_camera'] = df['Rear_camera'].str.split('+',expand=True)[0]

df['Rear_camera'] = df['Rear_camera'].str.split('MP',expand=True)[0]

df['Battery_Capacity'] = df['Battery_Capacity'].str.split(' ',expand=True)[0]

df['Front_camera'] = df['Front_camera'].str.split('MP',expand=True)[0]
```

```python
df
```

```python
#Checking the cardinality of each feature.
for i in df.columns:
    print(df[i].value_counts())
```

From above we see that in columns `RAM` and `ROM` there are values with MB. This could be because while scarping the websites, may be from few pages the recommended products data got scrpaed too. Let's deal with them now.

```python
ram_mask = (df['RAM'] == '128 MB') | (df['RAM'] == '32 MB') | (df['RAM'] == '16 MB') | (df['RAM'] == '10 MB') | (df['RAM'] == '4 MB') | (df['RAM'] == '0 MB')
df = df.drop(index=df[ram_mask].index)
```

```python
rom_mask = (df['ROM'] == '128 MB') | (df['ROM'] == '48 MB') | (df['ROM'] == '32 MB') | (df['ROM'] == '16 MB') | (df['ROM'] == '8 MB') |(df['ROM'] == '4 MB') | (df['ROM'] == '0 MB')
df = df.drop(index=df[rom_mask].index)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)].sort_values('Product_name')
```

There are duplicates in above dataframe means same products were repeating in different pages. This lead to the duplicates that we see above.

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.isna().sum()
```

This is obvious, there are nulls in most of the columns as the pattern that was set to scrape the web wasn't applicable for every product. Let's assume the most common feature that the nowadays smartphones are providing and fill the nulls with that

```python
print(df['Front_camera'].mode())
print(df['RAM'].mode())
print(df['Rear_camera'].mode())
print(df['Battery_Capacity'].mode())
print(df['Display_size'].median())
print(df['Processor'].mode())
```

Above are the features that are most commonly see in the smartphones nowadays

```python
df['RAM'] = df['RAM'].fillna('8')

df['Front_camera'] = df['Front_camera'].fillna('16')

df['Rear_camera'] = df['Rear_camera'].fillna('50')

df['Battery_Capacity'] = df['Battery_Capacity'].fillna('5000')

df['Display_size'] = df['Display_size'].fillna('6.58')

df['Processor'] = df['Processor'].fillna('Qualcomm Snapdragon 680')
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = df.astype({'Product_name':'string','After_Discount_Price':'int','RAM':'int','ROM':'int','Front_camera':'int',
           'Rear_camera':'int','Battery_Capacity':'int','Display_size':'float','Processor':'string','Brand':'category'})
```
