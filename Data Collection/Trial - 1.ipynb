```python
#Import required libraries
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
#Create empty list for each feature that you want to extract from the web
Product_name = []
After_Discount_Price = []
Product_Price = []
Description = []
RAM = []
ROM = []
Rear_camera = []
Front_camera = []
Display_size = []
Battery_Capacity = []
Processor_Type = []
Ratings_count = []
Reviews_count = []
Stars_out_of_5 = []
```

```python
#Define the URL
url = 'https://www.flipkart.com/search?q=mobiles+under+30000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page='
```

```python
#Fetching the data from URL 
open_url = requests.get(url)
print(open_url)
```

The above output `<Response [200]>` shows that the request has succeeded

```python
#Creating a object of BeautifulSoup
soup = BeautifulSoup(open_url.text,'lxml')
```

The above `soup` object contains the data of the whole page, which includes the ads and suggestions too which can mess with 
us in the further process. Now to avoid it, let's just take a section of page's data where the mobiles are listed. 

```python
#Creating a object named section, that has all the data we need.
section = soup.find('div',class_='_1YokD2 _3Mn1Gg')
```

```python
#Scrapping the names of the products.
names = section.find_all("div",class_='_4rR01T')
print(names)
```

```python
#Since the above scraped data isn't readable let's clean it a bit.
for i in names:
    Product_name.append(i.text)
print(len(Product_name))
Product_name
```

```python
#Scrapping the prices of the products after discount.
adp = section.find_all('div',class_='_30jeq3 _1_WHN1')
adp
```

```python
for i in adp:
    After_Discount_Price.append(i.text)
print(len(After_Discount_Price))
After_Discount_Price
```

```python
#Scrapping the Actual prices of the products.
price = section.find_all('div',class_='_3I9_wc _27UcVY')
price
```

```python
for i in price:
    Product_Price.append(i.text)
print(len(Product_Price))
Product_Price
```

```python
#Scrapping the description of each products.
des = section.find_all('div',class_='fMghEO')
des
```

```python
temp = []
for i in des:
    temp.append(i.text+' ')
print(len(temp))
temp
```

```python
#Segregating the description of each product into features as per our requirements
import re

for i in temp:
    # Defining regular expressions patterns for extracting information
    ram_pattern = r'(\d+ [GM]B) RAM'
    rom_pattern = r'(\d+ [GM]B) ROM'
    battery_pattern = r'(\d+ mAh) Battery'
    camera_pattern = r'(\d+MP \+ \d+MP) \| (\d+MP) Front Camera'
    processor_pattern = r'Battery(.+?)Processor'
    display_size_pattern = r'([\d.]+) inch'
    
    # Extract RAM information
    ram_match = re.search(ram_pattern,i)
    ram = ram_match.group(1) if ram_match else None

    # Extract ROM information
    rom_match = re.search(rom_pattern,i)
    rom = rom_match.group(1) if rom_match else None

    # Extract Battery information
    battery_match = re.search(battery_pattern,i)
    battery = battery_match.group(1) if battery_match else None

    # Extract Camera information
    camera_rear_match = re.search(camera_pattern,i)
    camera_rear = camera_rear_match.group(1) if camera_rear_match else None
    camera_front = camera_rear_match.group(2) if camera_rear_match else None

    # Extract Processor information
    processor_match = re.search(processor_pattern,i)
    processor = processor_match.group(1) if processor_match else None
    
    # Extract display size
    display_size_match = re.search(display_size_pattern, i)
    display_size = display_size_match.group(1) if display_size_match else None
    
    RAM.append(ram)
    ROM.append(rom)
    Battery_Capacity.append(battery)
    Rear_camera.append(camera_rear)
    Front_camera.append(camera_front)
    Processor_Type.append(processor)
    Display_size.append(display_size)

print(RAM)
print(ROM)
print(Battery_Capacity)
print(Rear_camera)
print(Front_camera)
print(Processor_Type)
print(Display_size)
```

```python
#Scrapping the ratings of the products.
ratings = section.find_all(['span'],class_='_2_R_DZ')
ratings
```

```python
temp_ratings = []
for i in ratings:
    temp_ratings.append(i.text)
    
temp_ratings
```

```python
for i in temp_ratings:
    # Defining the regular expression patterns to extract ratings count and reviews count
    ratings_pattern = r'([\d,]+) Ratings'
    reviews_pattern = r'([\d,]+) Reviews'

    # Extract ratings count
    ratings_match = re.search(ratings_pattern, i)
    ratings_count = ratings_match.group(1).replace(',', '') if ratings_match else None

    # Extract reviews count
    reviews_match = re.search(reviews_pattern, i)
    reviews_count = reviews_match.group(1).replace(',', '') if reviews_match else None
    
    Ratings_count.append(ratings_count)
    Reviews_count.append(reviews_count)

print(Ratings_count)
print(Reviews_count)
```

```python
##Scrapping the stars_out_of 5 of the each product.
stars = section.find_all('div',class_='_3LWZlK')
stars
```

```python
for i in stars:
    Stars_out_of_5.append(i.text)
Stars_out_of_5
```

```python
#Creating a dataframe to store the data that we just scraped from the web page.
temp_df = pd.DataFrame({'Product_name':Product_name,
             'Product_Price':Product_Price,
             'After_Discount_Price':After_Discount_Price,
             'RAM':RAM,
             'ROM':ROM,
             'Front_camera':Front_camera,
             'Rear_camera':Rear_camera,
             'Battery_Capacity':Battery_Capacity,
             'Display_size':Display_size,
             'Processor':Processor_Type,
             'Users_Reviewed':Reviews_count,
             'Users_Rated':Ratings_count,
             'Stars_out_of_5':Stars_out_of_5})
```
